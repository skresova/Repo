```python
import pandas
data = pandas.read_csv ("kiwo.csv")
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df['Datum'] = pd.to_datetime(df['Datum'])
df['weekday'] = df['Datum'].dt.weekday
df['dayofyear'] = df['Datum'].dt.dayofyear
df['day_of_week'] = df['Datum'].dt.day_name()
print (df)
df.to_csv('kiwo_updated.csv')
```
