# Merging Umsatzdaten with test dataset from Kaggle 

```python
test = pandas.read_csv ("test.csv")
data.to_csv("test_1.csv")
test
```

# Convert Datum in test dataset into timestamp

```python
test["Datum"]=pd.to_datetime(test["Datum"])
data2["Datum"]=pd.to_datetime(data2["Datum"])
data1["Datum"]=pd.to_datetime(data1["Datum"])
data["Datum"]=pd.to_datetime(data["Datum"])
test
```

# Merge Umsatzdaten with test dataset

```python
import pandas as pd
from datetime import datetime

combined = pd.merge(data2,test, how = 'outer')
combined.to_csv("combined.csv")
print(combined)
```

# Merging combined dataset (data2 + test) and data1 and data

```python
import pandas as pd
from datetime import datetime

combined_1 = pd.merge(combined,data1, on='Datum',how = 'left')
combined_1.to_csv("combined_1.csv")

valid = pd.merge(combined_1,data, on='Datum',how = 'left')
valid.to_csv("valid.csv")
print(valid)
```

# Add new columns (day of week, day of year)

```python
merged_data = pd.DataFrame(valid)
merged_data['Datum'] = pd.to_datetime(merged_data['Datum'])
merged_data['weekday'] = merged_data['Datum'].dt.weekday
merged_data['dayofyear'] = merged_data['Datum'].dt.dayofyear
merged_data['day_of_week'] = merged_data['Datum'].dt.day_name()
merged_data["month"] = merged_data['Datum'].dt.month
merged_data["year"] = merged_data['Datum'].dt.year
merged_data["weekofyear"] = merged_data['Datum'].dt.isocalendar().week
merged_data["quarter"] = merged_data['Datum'].dt.quarter
merged_data["is_weekend"] = merged_data['weekday'].apply(lambda x: 1 if x >=5 else 0)
merged_data["is_month_start"] = merged_data['Datum'].dt.is_month_start.astype(int)
merged_data["is_month_end"] = merged_data['Datum'].dt.is_month_end.astype(int)
merged_data["is_year_start"] = merged_data['Datum'].dt.is_year_start.astype(int)
merged_data["is_year_end"] = merged_data['Datum'].dt.is_year_end.astype(int)
merged_data["days_in_month"] = merged_data['Datum'].dt.days_in_month
merged_data["weekofmonth"] = merged_data['Datum'].apply(lambda d: (d.day - 1) // 7 + 1)
merged_data["is_leap_year"] = merged_data['Datum'].dt.is_leap_year.astype(int)
                                                          
merged_data.to_csv('merged_data.csv')
print (merged_data)
```

# Sort based on Warengruppe column

```python
sorted_merged_data= merged_data.sort_values(by=["Warengruppe", "Datum"])
sorted_merged_data.to_csv("sorted_merged_data.csv")
sorted_merged_data
```
