# Read data

```python
import pandas
all_data = pandas.read_csv ("all_data_holidays_events.csv")
all_data.to_csv("all_data_Svetlana.csv")
all_data
```

# Creating a new column Feiertag_Bereich_enc
# Feiertage: Feiertag_Bereich

```python
import numpy as np
all_data['Feiertag_Bereich_enc'] = np.where(all_data['Feiertag_Bereich'] == "bundesweit", 1, 0)
all_data
```

# One hot Encoding of the column Feiertag_Bereich_enc

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

all_data_encoded = pd.get_dummies(all_data, columns=['Feiertag_Bereich_enc'], drop_first=False)
all_data_encoded.to_csv("all_data_encoded.csv")
all_data_encoded
```

# Create a new column Feiertag_Bundesland_enc

```python
all_data_encoded['Feiertag_Bundesland_enc'] = np.where(all_data_encoded['Feiertag_Bundesland'] == "Schleswig-Holstein", 1, 0)
all_data_encoded
```

# One-hot encoding for the column Feiertag_Bundesland_enc

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

all_data_encoded_1 = pd.get_dummies(all_data_encoded, columns=['Feiertag_Bundesland_enc'], drop_first=False)
all_data_encoded_1.to_csv("all_data_encoded_1.csv")
all_data_encoded_1
```

# Create a new column based on the column Feiertag

```python
all_data_encoded_1["Feiertag_enc"] = all_data_encoded_1['Feiertag'].replace({'Tag der Deutschen Einheit': 1, 'Ostermontag': 2, 'Christi Himmelfahrt': 3, 
                                                                             "Pfingstmontag": 4, "Tag der Arbeit": 5}, inplace=False)
all_data_encoded_1.to_csv("all_data_encoded_2.csv")
all_data_encoded_1
```

# One-hot encoding of the column Feiertag_enc

```python
all_data_encoded_3 = pd.get_dummies(all_data_encoded_1, columns=['Feiertag_enc'], drop_first=False)
all_data_encoded_3.to_csv("all_data_encoded_3.csv")
all_data_encoded_3
```

# Creating a new numerical column for the column Wochentag

```python
all_data_encoded_3["Wochentag_enc"] = all_data_encoded_3['Wochentag'].replace({'Montag': 1, 'Dienstag': 2, 'Mittwoch': 3, 
                                                                             "Donnerstag": 4, "Freitag": 5, "Samstag":6, "Sonntag":7}, inplace=False)
all_data_encoded_3.to_csv("all_data_encoded_3.csv")
all_data_encoded_3
```

# One-hot encoding Wochentag_enc

```python
all_data_encoded_4 = pd.get_dummies(all_data_encoded_3, columns=['Wochentag_enc'], drop_first=False)
all_data_encoded_4.to_csv("all_data_encoded_4.csv")
all_data_encoded_4
```

# One-hot encoding for column Monat

```python
all_data_encoded_5 = pd.get_dummies(all_data_encoded_4, columns=['Monat'], drop_first=False)
all_data_encoded_5.to_csv("all_data_encoded_5.csv")
all_data_encoded_5
```

# One-hot encoding for column Jahr

```python
all_data_encoded_6 = pd.get_dummies(all_data_encoded_5, columns=['Jahr'], drop_first=False)
all_data_encoded_6.to_csv("all_data_encoded_6.csv")
all_data_encoded_6
```

# One-hot encoding for the column Warengruppe

```python
all_data_encoded_7 = pd.get_dummies(all_data_encoded_6, columns=['Warengruppe'], drop_first=False)
all_data_encoded_7.to_csv("all_data_encoded_7.csv")
all_data_encoded_7
```

# Create a new column with saison

```python
all_data_encoded_7["Datum"]=pd.to_datetime(all_data_encoded_7["Datum"])
all_data_encoded_7['Monat'] = all_data_encoded_7['Datum'].dt.month

all_data_encoded_7["Saison"] = all_data_encoded_7['Monat'].replace({12: "winter", 1: "winter", 2: "winter", 3: "spring", 4:"spring", 5:"spring", 
                                                                             6: "summer", 7: "summer", 8: "summer", 9: "autumn", 10: "autumn", 11: "autumn"}, 
                                                                             inplace=False)
all_data_encoded_7.to_csv("all_data_encoded_7.csv")
all_data_encoded_7
```

# Create a new column with numbers for the column Saison

```python
all_data_encoded_7["Saison_enc"] = all_data_encoded_7['Saison'].replace({"winter": 1, "spring": 2, "summer": 3, "autumn":4}, inplace=False)
all_data_encoded_7.to_csv("all_data_encoded_7.csv")
all_data_encoded_7
```

# One-hot encoding for the column Saison_enc

```python
all_data_encoded_8 = pd.get_dummies(all_data_encoded_7, columns=['Saison_enc'], drop_first=False)
all_data_encoded_8.to_csv("all_data_encoded_8.csv")
all_data_encoded_8
```
