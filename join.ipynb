# --- Merging files with os.path.join

```python
# importing libraries
import pandas as pd
import glob
import os

# merging the files
joined_files = os.path.join("data*.csv")

# A list of all joined files is returned
joined_list = glob.glob(joined_files)

# Finally, the files are joined
join_1 = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
join_1.to_csv("join_1.csv")
print(join_1)
```

```python
df_join = pd.DataFrame(join_1)
df_join['Datum'] = pd.to_datetime(df_join['Datum'])
df_join['weekday'] = df_join['Datum'].dt.weekday
df_join['dayofyear'] = df_join['Datum'].dt.dayofyear
df_join['day_of_week'] = df_join['Datum'].dt.day_name()
df_join.to_csv('df_join.csv')
print (df_join)
```
