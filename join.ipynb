```python
# importing libraries
import pandas as pd
import glob
import os

# merging the files
joined_files = os.path.join("data*.csv")

# A list of all joined files is returned
joined_list = glob.glob(joined_files)

# Finally, the files are joined
join_1 = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
join_1.to_csv("join_1.csv")
print(join_1)
```

```python
df_join = pd.DataFrame(join_1)
df_join['Datum'] = pd.to_datetime(df_join['Datum'])
df_join['weekday'] = df_join['Datum'].dt.weekday
df_join['dayofyear'] = df_join['Datum'].dt.dayofyear
df_join['day_of_week'] = df_join['Datum'].dt.day_name()
df_join.to_csv('df_join.csv')
print (df_join)
```

# It worked!!!! Next trial for merging

```python
import pandas as pd
from datetime import datetime

combinecsv_2 = pd.merge(data1,data2, on='Datum',how = 'left')
combinecsv_2.to_csv("combined_2.csv")

validcsv_2 = pd.merge(combinecsv_2,data, on='Datum',how = 'left')
validcsv_2.to_csv("valid_2.csv")
print(validcsv_2)
```

# Add new columns (day of week, day of year)

```python
merged_data = pd.DataFrame(validcsv_2)
merged_data['Datum'] = pd.to_datetime(merged_data['Datum'])
merged_data['weekday'] = merged_data['Datum'].dt.weekday
merged_data['dayofyear'] = merged_data['Datum'].dt.dayofyear
merged_data['day_of_week'] = merged_data['Datum'].dt.day_name()
merged_data.to_csv('merged_data.csv')
print (merged_data)
```
