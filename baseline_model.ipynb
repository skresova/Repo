# Read data

```python
import pandas
all_data = pandas.read_csv ("all_data.csv")
all_data
```

# Keep the id column

```python
all_data.set_index('id', inplace=True)
all_data
```

# Split dataset into training, validation and test datasets

```python
# Define your date thresholds
train_end_date = '2017-07-31'
validation_end_date = '2018-07-31'

# Convert to datetime if not already
import pandas as pd
all_data['Datum'] = pd.to_datetime(all_data['Datum'])

# Split the data based on the date thresholds
train_data_all = all_data[all_data['Datum'] <= train_end_date]
validation_data_all = all_data[(all_data['Datum'] > train_end_date) & (all_data['Datum'] <= validation_end_date)]
test_data_all = all_data[all_data['Datum'] > validation_end_date]

# Check the dimensions of the datasets
print("Training dataset dimensions:", train_data_all.shape)
print("Validation dataset dimensions:", validation_data_all.shape)
print("Test dataset dimensions:", test_data_all.shape)

# Save the datasets to CSV files
train_data_all.to_csv("train_data_all.csv", index=False)
validation_data_all.to_csv("validation_data_all.csv", index=False)
test_data_all.to_csv("test_data_all.csv", index=False)
```

# Training linear model

```python
import pandas as pd
import statsmodels.formula.api as smf

# Fit the linear model
lin_mod = smf.ols('Umsatz ~ Bewoelkung + Temperatur + Windgeschwindigkeit + C(Produktgruppe) + Wettercode + ' \
'Veranstaltung_Kieler_Woche + Monat + Jahr', data=train_data_all).fit()

# Print the summary
print(lin_mod.summary())
```

# Predict the Umsatz

```python
predictions_all = lin_mod.predict(train_data_all)
print(predictions_all)
```

# Drop NA from validation_data_all dataset

```python
validation_data_all=validation_data_all.dropna()
validation_data_all
```

```python
predictions_valid_all = lin_mod.predict(validation_data_all)
predictions_valid_all.to_csv("predictions_validation_all.csv")
predictions_valid_all
```

# Calculate RMSE & MSE

```python
from sklearn.metrics import mean_squared_error # Import mean_squared_error from sklearn.metrics
import math # Import math library

# Calculate the mean squared error using mean_squared_error from sklearn.metrics
mse_valid_all = mean_squared_error(validation_data_all[["Umsatz"]], predictions_valid_all)
 
# Raise the mean squared error to the power of 0.5
rmse_valid_all = (mse_valid_all)**(1/2)
# Print the RMSE
rmse_valid_all
```

# Impute missing values in the test_data_all

```python
# Fill missing values with the median
test_data_all['Wettercode'] = test_data_all['Wettercode'].fillna(test_data_all['Wettercode'].median())
test_data_all['Temperatur'] = test_data_all['Temperatur'].fillna(test_data_all['Temperatur'].median())
test_data_all["Bewoelkung"] =  test_data_all["Bewoelkung"].fillna(test_data_all["Bewoelkung"].median())
test_data_all["Windgeschwindigkeit"] =  test_data_all["Windgeschwindigkeit"].fillna(test_data_all["Windgeschwindigkeit"].median())
test_data_all["KielerWoche"] =  test_data_all["KielerWoche"].fillna(test_data_all["KielerWoche"].median())

test_data_all
```

```python
predictions_test_all = lin_mod.predict(test_data_all)
predictions_test_all.to_csv("predictions_test_all_Team_5.csv")
predictions_test_all
```
